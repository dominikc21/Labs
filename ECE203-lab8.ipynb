```python
import matplotlib.pyplot as plt
```

```python
#data for diode characteristic curve
rev_v = [0, -0.5, -1, -1.5, -2] #V
rev_i = [-0.873, -0.873, -0.8707, -0.873, -0.8707] #uA

for_v = [0.542, 0.5737, 0.5916, 0.6063, 0.6172] #V
for_i = [250, 500, 750, 1000, 1250] #uA

#data for biased positive shunt clipper
v_in = [0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4]
v_out = [0.4396, 0.896, 1.351, 1.805, 2.129, 2.196, 2.26, 2.345]
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.plot(rev_v, rev_i, label="Rev. Biased Current", marker='o')
ax1.plot(for_v, for_i, label="For. Biased Current", marker='o')
ax1.set_xlabel("Diode Voltage (V)")
ax1.set_ylabel("Diode Current (uA)")
ax1.set_title("Forward and reverse biased diode current vs. diode voltage")

plt.legend(bbox_to_anchor=[1,1])
plt.savefig("ECE203_lab8_plot1", bbox_inches="tight")
```

```python
fig2 = plt.figure()
ax2 = fig2.add_subplot(111)
ax2.plot(v_in, v_out, label='V_out', marker='o')
ax2.set_xlabel("Voltage in (V)")
ax2.set_ylabel("Voltage out (V)")
ax2.set_title("V_out vs V_in")

plt.legend(bbox_to_anchor=[1, 1])
plt.savefig("ECE203_lab8_plot2", bbox_inches="tight")
```
