```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel(r'/Users/dominikculo/Downloads/ECE203_lab6_data.xlsx', usecols='A,B,C,D,F,G,H,I,K,M,N,P,Q,R,S')

#low pass RC
freq_low_RC = pd.DataFrame(df, columns=['Freq(low RC)'])
Vin_low_RC = pd.DataFrame(df, columns=['Vin(low RC)'])
Vout_low_RC = pd.DataFrame(df, columns=['Vout(low RC)'])
att_low_RC = pd.DataFrame(df, columns=['Vout/Vin(low RC)'])

#high pass RC
freq_high_RC = pd.DataFrame(df, columns=['Freq(high RC)'])
Vin_high_RC = pd.DataFrame(df, columns=['Vin(high RC)'])
Vout_high_RC = pd.DataFrame(df, columns=['Vout(high RC)'])
att_high_RC = pd.DataFrame(df, columns=['Vout/Vin(high RC)'])

#low pass RL
freq_low_RL = pd.DataFrame(df, columns=['Freq(low RL)'])
Vin_low_RL = pd.DataFrame(df, columns=['Vin(low RL)'])
Vout_low_RL = pd.DataFrame(df, columns=['Vout(low RL)'])
att_low_RL = pd.DataFrame(df, columns=['Vout/Vin(low RL)'])

#band pass
freq_band = pd.DataFrame(df, columns=['Freq(band)'])
Vin_band = pd.DataFrame(df, columns=['Vin(band)'])
Vout_band = pd.DataFrame(df, columns=['Vout(band)'])
att_band = pd.DataFrame(df, columns=['Vout/Vin(band'])
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(111)
ax1.plot(freq_low_RC, att_low_RC, label='Low-pass RC')
ax1.plot(freq_high_RC, att_high_RC, label='High-pass RC')

plt.legend(bbox_to_anchor=[1, 1])
ax1.set_yscale('log')
ax1.set_xscale('log')
ax1.set_xlabel('Frequency (Hz)')
ax1.set_ylabel('Attenuation (%)')
ax1.set_title('RC Low-pass and High-pass Filters')

plt.savefig('ECE203_lab6_fig1', bbox_inches='tight')
```

```python
fig2 = plt.figure()

ax2 = fig2.add_subplot(111)
ax2.plot(freq_low_RL, att_low_RL, label='Low-pass RL')

plt.legend(bbox_to_anchor=[1, 1])
ax2.set_yscale('log')
ax2.set_xscale('log')
ax2.set_xlabel('Frequency (Hz)')
ax2.set_ylabel('Attenuation (%)')
ax2.set_title('RL Low-pass Filter')

plt.savefig('ECE203_lab6_fig2', bbox_inches='tight')
```

```python
fig3 = plt.figure()

ax3 = fig3.add_subplot(111)
ax3.plot(freq_band, att_band, label='Band-pass')

plt.legend(bbox_to_anchor=[1, 1])
ax3.set_yscale('log')
ax3.set_xscale('log')
ax3.set_xlabel('Frequency (Hz)')
ax3.set_ylabel('Attenuation (%)')
ax3.set_title('RLC Band-pass Filter')

plt.savefig('ECE203_lab6_fig3', bbox_inches='tight')
```
