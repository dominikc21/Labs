```python
import pandas as pd
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
df = pd.read_excel(r'/Users/dominikculo/downloads/ECE203_lab7.xlsx', usecols='A,B,C,E,F,G,I,J,K,M,N,O')

#inverting amplifier - changing Rf
Rf1 = pd.DataFrame(df, columns=['Rf1']).dropna()
Gain1 = pd.DataFrame(df, columns=['Gain1']).dropna()
Ideal1 = pd.DataFrame(df, columns=['Ideal1']).dropna()

# // // - increasing Vin
Vin2 = pd.DataFrame(df, columns=['Vin2']).dropna()
Gain2 = pd.DataFrame(df, columns=['Ideal2']).dropna()

# // // - increasing f
freq = pd.DataFrame(df, columns=['freq']).dropna()
Gain3 = pd.DataFrame(df, columns=['Gain3']).dropna()
Ideal3 = pd.DataFrame(df, columns=['Ideal3']).dropna()

#non-inverting amplifier
Rf4 = pd.DataFrame(df, columns=['Rf4']).dropna()
Gain4 = pd.DataFrame(df, columns=['Gain4']).dropna()
Ideal4 = pd.DataFrame(df, columns=['Ideal4']).dropna()

frequency = [100, 300, 1000, 3000, 10000, 30000, 100000, 300000, 1000000, 3000000]
Gain3 = [4.68, 4.68, 4.67, 4.6, 4.65, 4.48, 1.91, 0.629, 0.187, 0.123]
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(111)
ax1.plot(Rf1, Gain1, marker='o', label='Measured gain')
ax1.plot(Rf1, Ideal1, marker='o', label='Ideal gain')

plt.legend(bbox_to_anchor=[1, 1])
ax1.set_xlabel('Resistance(Ω)')
ax1.set_ylabel('Gain')
ax1.set_title('Inverting Amplifier - Varying Rf')

plt.savefig('ECE203_lab7_fig1', bbox_inches='tight')
```
